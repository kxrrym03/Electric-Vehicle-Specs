### Handling null values in my dataset
Handle null values in the dataset by either dropping rows with nulls or imputing missing values, and then verify that there are no remaining null values.

### Identify null values

### Subtask:
Analyze the null values in each column and determine the best strategy to handle them.


**Reasoning**:
Analyze the null values in each column to determine the best strategy for handling them by calculating the percentage of missing values for each column with missing data and displaying them.



```python
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_percentage = (missing_values / len(df)) * 100

missing_info = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage': missing_percentage})
display(missing_info)
```

## Handle null values

### Subtask:
Depending on the column and the number of nulls, either drop the rows with nulls or impute the missing values.


**Reasoning**:
Handle null values by dropping the 'number_of_cells' column and imputing missing values in other columns as specified in the instructions.



```python
df = df.drop('number_of_cells', axis=1)

for col in ['torque_nm', 'fast_charging_power_kw_dc', 'towing_capacity_kg']:
    df[col].fillna(df[col].median(), inplace=True)

for col in ['model', 'fast_charge_port', 'cargo_volume_l']:
    df[col].fillna(df[col].mode()[0], inplace=True)

df.isnull().sum()
```

## Verify null values

### Subtask:
Verify that there are no remaining null values in the dataset.


**Reasoning**:

### Visualizations

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='brand')
plt.title('Brand Distribution')
plt.xticks(rotation=90)
plt.xlabel('Brand')
plt.show()
```

```python
max_speed_by_brand = df.groupby('brand')['top_speed_kmh'].max()
display(max_speed_by_brand)
```

```python
top_10_brands_speed = max_speed_by_brand.sort_values(ascending=False).head(10)
display(top_10_brands_speed)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_brands_speed.index, y=top_10_brands_speed.values)
plt.title('Top 10 Brands by Maximum Top Speed')
plt.xlabel('Brand')
plt.ylabel('Maximum Top Speed (km/h)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
car_body_type_counts = df['car_body_type'].value_counts()
display(car_body_type_counts)
```

```python
labels = car_body_type_counts.index
sizes = car_body_type_counts.values
```

```python
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Car Body Types')
plt.axis('equal')
plt.show()
```
